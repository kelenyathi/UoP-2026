Exercise 1 - This program asks user to enter a number and checks if it is a prime number

```python
num = int(input("Enter a number: "))

is_prime = True

if num <= 1:
    is_prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
        print ("THIS IS A PRIME NUMBER")
else:
    print("THIS IS NOT A PRIME NUMBER")  
```

Exercise 5 - Function that accepts any iterable (list, tuple, set, range, string) of numbers.

```python
def find_extremes(data):
    #this will return the smallest and largest numbers as a pair 
    return min(data), max(data)

my_nums = [10,40,1,90,32]
low, high = find_extremes(my_nums)

print(f"Min value is: {low}")
print(f"Max value is: {high}")
```
