Exercise 1 - This program asks user to enter a number and checks if it is a prime number

```python
num = int(input("Enter a number: "))

is_prime = True

if num <= 1:
    is_prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
        print ("THIS IS A PRIME NUMBER")
else:
    print("THIS IS NOT A PRIME NUMBER")  
```

Exercise 5 - Function that accepts any iterable (list, tuple, set, range, string) of numbers.

```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression

tips=sns.load_dataset('tips')
X=tips[['total_bill']]
y=tips['tip']
model=LinearRegression()
model.fit(X,y)

X_new = pd.DataFrame([[34.00]], columns=['total_bill'])
prediction = model.predict(X_new)

print (f"Predicted tip for £34 bill:, £{prediction[0]:.2f}")
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
sns.regplot(
    data=tips,
    x='total_bill', 
    y='tip', 
    scatter_kws={'alpha':0.3},
    line_kws={'color':'red'}
)

plt.title("bill vs tip")
plt.xlabel("Total bill")
plt.ylabel("Tip amount")
plt.show()
```
