Exercise 1 - This program asks user to enter a number and checks if it is a prime number

```python
num = int(input("Enter a number: "))

is_prime = True

if num <= 1:
    is_prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
        print ("THIS IS A PRIME NUMBER")
else:
    print("THIS IS NOT A PRIME NUMBER")  
```

Exercise 8 - Code ‍‍to ‍‍define ‍‍and‍ run‍ the ‍‍function‍ solving ‍‍the‍ series

```python
#importing maths library
import math 
def series (x,N):
    total=0

    for i in range (1, N +1):
        term=(x*(5**i)**2)/(i+math.factorial(i+1))
        total +=term
    return total
try:
    val_x = int(input("Enter x:"))
    val_N = int(input("Enter N:"))

    result=series(val_x, val_N)
    print("Result:", result)

except ValueError:
    #message shown when whole number is not entered
    print("Wrong Input. Enter a whole number")
```

Exercise 9 - This loads an image and creates horizontal and vertical reflections

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

#image loacation provided
img= Image.open ("/workspaces/UoP-2026/jupyter_notebook/flowers.jpeg")
img_array = np.array (img)

#direction image should be flipped
horizontal = np.fliplr(img_array)
vertical = np.flipud (img_array)

#axis of image visualisation
plt.imshow (horizontal)
plt.title ("Horizontal Reflection")
plt.show()

plt.imshow(vertical)
plt.title("Vertical Reflection")
plt.show()
```

```python
#imported data sets
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
#loaded data sets
tips=sns.load_dataset('tips')
X=tips[['total_bill']]
y=tips['tip']
model=LinearRegression()
model.fit(X,y)

X_new = pd.DataFrame([[34.00]], columns=['total_bill'])
prediction = model.predict(X_new)

print (f"Predicted tip for £34 bill:, £{prediction[0]:.2f}")
```

```python
#imported data sets
import matplotlib.pyplot as plt 
import seaborn as sns
sns.regplot(
    data=tips,
    x='total_bill', 
    y='tip', 
    scatter_kws={'alpha':0.3},
    line_kws={'color':'red'}
)

plt.title("bill vs tip")
plt.xlabel("Total bill")
plt.ylabel("Tip amount")
plt.show()
```
